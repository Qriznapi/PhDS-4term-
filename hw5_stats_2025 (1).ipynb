**3.** Реализуйте функцию вычисления выборки оценок $\theta^*_1, \ldots \theta^*_B$ для оценки $\widehat \theta$ методом бутстрепа. Функция должна принимать на вход выборку и количество бутстрепных выборок $B$.

Для ускорения расчетов можете использовать broadcasting (не является обязательным требованием). Учите, что в таком случае внутри функции у вас могут получиться очень большие матрицы, из-за которых может кончиться оперативная память.

```python
def bootstrap(sample, B=100_000):
    '''
    Считает бутстрепные оценки для исходной выборки

    :param sample: исходная выборка
    :param B: количество бутстрепных
    :return bootstrap_estimations: оценки по бутстрепным выборкам
    '''
    n = len(sample)
    sample = np.asarray(sample)
    return np.array([np.mean(np.random.choice(sample, n)) for _ in range(B)])
```

**4.** Реализуйте три типа бутстрепных доверительных интервалы в виде функций, принимающих на вход выборку оценок, полученных с помощью бутстрепа. Реализовывать вычисление бутстрепных интервалов для подвыборок размера от 1 до n не нужно, только для всей выборки.

```python
def bootstrap_normal_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу нормального бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''
    se = np.std(bootstrap_estimations, ddof=1)
    z = sps.norm.ppf((1 + alpha)/2)
    left = theta - z * se
    right = theta + z * se
    return left, right


def bootstrap_central_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу центрального бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''
    deviations = bootstrap_estimations - theta
    q = (1 + alpha)/2
    left = theta - np.quantile(np.abs(deviations), q)
    right = theta + np.quantile(np.abs(deviations), q)
    return left, right


def bootstrap_quantile_confidence_interval(theta, bootstrap_estimations, alpha=0.95):
    '''
    Считает левую и правую границу квантильного бутстрепного интервала

    :param theta: оценка параметра
    :param bootstrap_estimations: массив бутстрепных оценок
    :return left: левая граница бутстрепного интервала
    :return right: правая граница бутстрепного интервала
    '''
    conf_low = (1 - alpha)/2
    conf_high = (1 + alpha)/2
    left = np.quantile(bootstrap_estimations, conf_low)
    right = np.quantile(bootstrap_estimations, conf_high)
    return left, right
```

**5.** Для каждого признака постройте бутстрепные доверительные интервалы для $\theta = \mathsf{E} X_1$ и сравните их.

```python
data = pd.DataFrame(index = ['normal', 'central', 'quantile', 'theta'], columns=df.columns)
```

```python
for feature in df.columns:
    sample = df[feature].values
    theta = np.mean(sample)
    bootstrap_estimations = bootstrap(sample, B=10000)

    normal_ci = bootstrap_normal_confidence_interval(theta, bootstrap_estimations)
    central_ci = bootstrap_central_confidence_interval(theta, bootstrap_estimations)
    quantile_ci = bootstrap_quantile_confidence_interval(theta, bootstrap_estimations)

    data.loc['normal', feature] = f"({normal_ci[0]:.4f}, {normal_ci[1]:.4f})"
    data.loc['central', feature] = f"({central_ci[0]:.4f}, {central_ci[1]:.4f})"
    data.loc['quantile', feature] = f"({quantile_ci[0]:.4f}, {quantile_ci[1]:.4f})"
    data.loc['theta', feature] = f"{theta:.4f}"

    data.loc['interval_length_normal', feature] = normal_ci[1] - normal_ci[0]
    data.loc['interval_length_central', feature] = central_ci[1] - central_ci[0]
    data.loc['interval_length_quantile', feature] = quantile_ci[1] - quantile_ci[0]

print("Таблица с результатами:")
display(data)
```

Визуализируйте бутстрепные интервалы для каждого признака. Для этого сгенерируйте выборку $X_1, ... X_{N}, N = 100$ и постройте график доверительных интервалов уровня доверия $0.95$, вычисленных для всех подвыборок размера $n$ вида $X_1, ... X_n$, $1 \le n \le 100$, используя написанную ниже функцию

```python
def draw_confidence_interval(
    left,
    right,
    estimation=None,
    sample=None,
    ylim=(None, None),
    estim_label = '',
    sample_label='',
    color=None,
    interval_label=None
):
    '''
    Рисует доверительный интервал и оценку в зависимости от размера выборки.

    :param left: левые границы интервалов (в зависимости от n)
    :param right: правые границы интервалов (в зависимости от n)
    :param estimation: оценки (в зависимости от n)
    :param sample: выборка
    :param ylim: ограничение вертикальной оси
    :param estim_label: подпись к оценке
    :param sample_label: подпись к выборке
    :param color: цвет, которым будет отображен доверительный интервал
    '''

    plt.figure(figsize=(7, 5))


    left = np.asarray(left)
    right = np.asarray(right)

    if estimation is not None:
        x = np.arange(1, len(estimation)+1)
    else:
        x = np.arange(1, len(left)+1)

    if color is None:
        color = 'blue'

    plt.fill_between(x, left, right, color=color, alpha=0.2,
                    label=interval_label if interval_label else 'Доверительный интервал')
    plt.plot(x, left, color=color, linestyle='--', linewidth=1)
    plt.plot(x, right, color=color, linestyle='--', linewidth=1)

    if estimation is not None:
        estimation = np.asarray(estimation)
        plt.plot(x, estimation, color='red', marker='o', markersize=2,
                label=estim_label if estim_label else 'Оценка параметра')

    plt.xlabel('Размер выборки')
    plt.ylabel('Значение параметра')
    plt.title('Доверительные интервалы в зависимости от размера выборки')
    plt.ylim(ylim)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()

    plt.tight_layout()
    plt.show()

    return
```

**Решение:**

```python
features = ['E1', 'E2', 'M']

for feature in features:
    feature_data = df[feature].values

    lowers, uppers, estimates = [], [], []
    for n in range(1, 100):
        sub_sample = feature_data[:n]
        bootstrap_estimations = bootstrap(sub_sample, B=100)
        lower, upper = bootstrap_normal_confidence_interval(theta, bootstrap_estimations)
        lowers.append(lower)
        uppers.append(upper)
        estimates.append(np.mean(sub_sample))

    draw_confidence_interval(
        left=lowers,
        right=uppers,
        estimation=estimates,
        sample=feature_data,
        estim_label='Среднее значение',
        sample_label='Распределение данных',
        color='green',
        interval_label='нормальный доверительный интервал',
        ylim=(np.min(feature_data)*0.9, np.max(feature_data)*1.1)
    )

    feature_data = df[feature].values

    lowers, uppers, estimates = [], [], []
    for n in range(1, 100):
        sub_sample = feature_data[:n]
        bootstrap_estimations = bootstrap(sub_sample, B=100)
        lower, upper = bootstrap_central_confidence_interval(theta, bootstrap_estimations)
        lowers.append(lower)
        uppers.append(upper)
        estimates.append(np.mean(sub_sample))

    draw_confidence_interval(
        left=lowers,
        right=uppers,
        estimation=estimates,
        sample=feature_data,
        estim_label='Среднее значение',
        sample_label='Распределение данных',
        color='green',
        interval_label='центральный доверительный интервал',
        ylim=(np.min(feature_data)*0.9, np.max(feature_data)*1.1)
    )

    feature_data = df[feature].values

    lowers, uppers, estimates = [], [], []
    for n in range(1, 100):
        sub_sample = feature_data[:n]
        bootstrap_estimations = bootstrap(sub_sample, B=100)
        lower, upper = bootstrap_quantile_confidence_interval(theta, bootstrap_estimations)
        lowers.append(lower)
        uppers.append(upper)
        estimates.append(np.mean(sub_sample))

    draw_confidence_interval(
        left=lowers,
        right=uppers,
        estimation=estimates,
        sample=feature_data,
        estim_label='Среднее значение',
        sample_label='Распределение данных',
        color='green',
        interval_label='квантильный доверительный интервал',
        ylim=(np.min(feature_data)*0.9, np.max(feature_data)*1.1)
    )
```

Квантильный ДИ сильно больше других при малых значениях выборки. Возможно стоит использовать центральный ДИ при малых значениях выборки и квантильный - при больших.
