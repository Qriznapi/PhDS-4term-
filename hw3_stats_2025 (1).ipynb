Эквивалентный асимптотичсекий доверительный интервал для параметра $\theta$ уровня доверия $1-\alpha$

На лекции вы узнали про p-value &mdash; это вероятность получить при справедливости $H_0$ такое значение статистики $t = T(x)$ или еще более экстремальное, то есть в случае двустороннего критерия
$$p(x) = \mathsf{P}_0(T(X) \geq|t|) + \mathsf{P}_0(T(X) \leq -|t|)$$
Для расчета можно использовать функции из библиотеки `scipy.stats`.

**2**. Оформите функцию подсчета статистики критерия, p-value и доверительного интервала

```python
def wald_test(sample, theta, estimation_theta, estimation_sigma, alternative='two_sided'):
    """
    param sample: реализация выборки
    param theta: истинное значение параметра
    param estimation_theta: оценка параметра
    param estimation_sigma: оценка асимптотической дисперсии
        оценки estimation_sigma
    param alternative: вид альтернативной гипотезы,
        может принимать одно из значений 'two_sided', 'less', 'greater'

    return statistic
    return p_value
    return conf_int - доверительный интервал
    """
    n = len(sample)
    se = np.sqrt(estimation_sigma) / np.sqrt(n)

    statistic = np.sqrt(n) * np.abs(estimation_theta - theta) / np.sqrt(estimation_sigma)

    if alternative == 'two_sided':
        p_value = 2 * sps.norm.sf(statistic)
        z = sps.norm.ppf(1 - 0.05/2)
        conf_int = (estimation_theta - z * se, estimation_theta + z * se)
    elif alternative == 'greater':
        p_value = sps.norm.sf(statistic)
        z = sps.norm.ppf(1 - 0.05)
        conf_int = (-np.inf, estimation_theta + z * se)
    else:
        p_value = sps.norm.cdf(statistic)
        z = sps.norm.ppf(1 - 0.05)
        conf_int = (estimation_theta - z * se, np.inf)

    return statistic, p_value, conf_int
```

**3**. Проверьте гипотезу с двусторонней альтернативой.

```python
Ex = np.mean(sample)
lambdag = 1 / Ex

sigmab = lambdag ** 2

stat, p_val, conf_int = wald_test(sample, 2, lambdag, sigmab, 'two_sided')

print(f"Статистика критерия: {stat:.3f}")
print(f"p-value: {p_val:.4f}")
print(f"Доверительный интервал: ({conf_int[0]:.3f}, {conf_int[1]:.3f})")
```

**Вывод:** значение p-value < 0.05, основная гипотеза отвергается, в чем мы уже убеждались, при этом 2 не лежит в доверительном интервале

**4**. Проверьте гипотезу с правосторонней альтернативой.

```python
Ex = np.mean(sample)
lambdag = 1 / Ex

sigmab = lambdag ** 2

stat, p_val, conf_int = wald_test(sample, 2, lambdag, sigmab, 'greater')

print(f"Статистика критерия: {stat:.3f}")
print(f"p-value: {p_val:.4f}")
print(f"Доверительный интервал: ({conf_int[0]:.3f}, {conf_int[1]:.3f})")
```

**Вывод:** значение p-value < 0.05 (даже еще меньше, тк сравнение с правой частью стало еще значимее), основная гипотеза отвергается, что логично, тк среднее находится правее гипотезы. Значение находится в ДИ, однако это не приводит к подтверждению гипотезы.

### Задача 3

На практике часто рассматривают асимптотические критерии, например, критерий Вальда. Такие критерии контролируют вероятность ошибки I рода на уровне $\alpha$ только в пределе, однако для небольших выборок она может сильно отличаться от $\alpha$, причем нередко в большую сторону. Поэтому для таких критериев возникает необходимость в вычислении значения вероятности ошибки I рода или *реального уровня значимости* при конечной выборке.

Зачастую посчитать точное значение реального уровня значимости довольно сложно, поэтому вместо него вычисляют оценку с помощью сэмплирования по методу Монте-Карло.

Пусть гипотеза $\mathsf{H}_0$ простая, то есть $\mathsf{H}_0\colon \mathsf{P} = \mathsf{P}_0$. Для ее проверки по конечной выборке $X_1, ..., X_n$ используется критерий $S$. **Опишите схему оценки реального уровня** значимости критерия $S$ методом Монте-Карло.


**Ответ:**

Вычисляем вероятность ошибки первого рода для каждой выборки, соответствующей нашей. Реальный уровень значимости определяется как среднее полученных вероятностей

Если гипотеза $\mathsf{H}_0$ сложная, то есть $\mathsf{H}_0\colon \mathsf{P} \in \mathscr{P}_0$, то можно повторить такую процедуру несколько раз для разных $\mathsf{P} \in \mathscr{P}_0$.

Рассмотрим альтернативную гипотезу $\mathsf{H}_1\colon \mathsf{P} \in \mathscr{P}_1$. **Опишите схему оценки мощности** критерия $S$ методом Монте-Карло.

**Ответ:**

Метод тот же, только теперь мы будем определять отвержение основаной гипотезы при помощи статистики критерия, а сама мощность определяется так же средним числом отклонений гипотезы по всем выборкам при верной альтернативе

Пусть $X_1, ..., X_n$ &mdash; выборка из пуассоновского распределения с параметром $\theta$. Рассмотрим гипотезы $\mathsf{H}_0\colon \theta = 1\ \ vs.\ \ \mathsf{H}_1\colon \theta \not= 1$ и критерий Вальда для их проверки.

**1.** **Оцените реальный уровень значимости** критерия для $ n = 50$.

```python
n = 50
theta0 = 1
alpha = 0.05
N = 2500

np.random.seed(42)
samples = np.random.poisson(lam=theta0, size=(N, n))

thetah = samples.mean(axis=1)

se = np.sqrt(thetah / n)
W = (thetah - theta0) / se

z_critical = sps.norm.ppf(1 - alpha/2)
reject = np.abs(W) > z_critical

# Оценка реального уровня значимости
alpha_real = reject.mean()
print(f"Реальный уровень значимости: {alpha_real:.4f}")
```

**Вывод:** реальный уровень значимости меньше теоретического для небольших размеров выборки

**2.** Пусть $n=50$. **Сравните мощности критериев**, соответствующих двусторонней, левосторонней и правосторонней альтернативам при $\theta=0.2$ и при $\theta=2$.

```python
n = 50
theta0 = 1
alpha = 0.05
N = 100000
thetas = [0.2, 2]

z_left = sps.norm.ppf(alpha)
z_right = sps.norm.ppf(1 - alpha)
z_two_sided = sps.norm.ppf(1 - alpha/2)

power_two_sided = []
power_left = []
power_right = []

for theta in thetas:
    samples = np.random.poisson(lam=theta, size=(N, n))

    theta_hats = samples.mean(axis=1) + 0.00001
    se = np.sqrt(theta_hats / n)
    W = (theta_hats - theta0) / se

    reject_two_sided = np.abs(W) > z_two_sided
    reject_left = W < z_left
    reject_right = W > z_right

    power_two_sided.append(reject_two_sided.mean())
    power_left.append(reject_left.mean())
    power_right.append(reject_right.mean())

for i in range(len(thetas)):
    print(f"θ = {thetas[i]}:")

    print(f"  Двусторонний: {power_two_sided[i]:.4f}")
    print(f"  Левосторонний: {power_left[i]:.4f}")
    print(f"  Правосторонний: {power_right[i]:.4f}")
    print()
```

**Вывод:** Мощности очень близки к 0 и 1 для соответствующих θ, что значит, что в первом случае правосторонний критерий не применим, во втором - левосторонний
