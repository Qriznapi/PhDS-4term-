### Задача 3


Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$. Проверьте эти оценки на несмещенность, повторив эксперимент с семинара.

Реализуйте две функции, каждая из которых на вход берет несколько выборок, а на выход выдает массив оценок (первая функция - $S^2$, вторая функция - $\frac{n}{n-1}S^2$)  для каждой выборки.  
*Примечание:* У матрицы, заданной в виде `numpy.array` есть метод `var` вычисляющий выборочную дисперсию.

```python
def estimate_s2(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    return X.var(axis=1)

def estimate_s2_corrected(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    return X.var(axis=1) * len(X[0]) / (len(X[0]) - 1)
```

Проверьте правильность своей реализации

```python
X = np.array([[1, 3, 3, 6], [2, 5, 1, 89]])
estimate_s2(X)

assert(sum(estimate_s2(X) != np.array([3.1875, 1399.6875])) == 0)
assert(sum(estimate_s2_corrected(X) != np.array((4.25, 1866.25))) == 0)
```

Мы хотим понять, являются ли оценки параметра $\theta$ смещенными, для этого нам нужно провести множество экспериментов (сгенерировать выборку много раз, так как при каждой генерации получаются разные числа).

Напишите функцию построения таких графиков в общем виде: на вход функция может получать любые распределения (`distributions`) и любые оценки (`estimators`).
Средние значения по экспериментам для каждой оценки нужно обозначить звездочками.

*Подсказка: получается, что для этой задачи достаточно просто взять функции из семинарского ноутбука?!*

```python
def est_plot(distribution, estimators, sample_size, sample_count):
    '''
    Построение графика разброса реализаций оценок и их средних значений.

    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    sample_size -- размер выборок
    sample_count -- количество генерируемых выборок
    '''
    X = distribution.rvs(size=(sample_count, sample_size))
    plt.figure(figsize=(15, 0.7*len(estimators)))
    for estimator, name, color, y in estimators:
        E = estimator(X)
        E = np.asarray(E)
        plt.scatter(E, np.zeros(sample_count) + y, alpha=0.1,
                        s=100, color=color, label=name)
        plt.scatter(E.mean(), y, marker='*', s=300,
                        color='w', edgecolors='black')

        plt.vlines(1, -1, 1, color='r')
        plt.title('Размер выборки = %d' % sample_size)
        plt.yticks([])
        plt.legend()
        plt.xlim((0.8, 1.2))
        plt.ylim((-0.1, 0.1 * len(estimators)))
```

Постройте три графика для разных размеров выборок (10, 100, 500).


```python
sample_size_list = (10, 100, 500)  # размеры выборок
sample_count = 500  # количество экспериментов

distribution = norm(loc=0, scale=1)

for sample_size in sample_size_list:
    est_plot(
        distribution=distribution,
        estimators=estimators,
        sample_size=sample_size,
        sample_count=sample_count
    )
```

Изучите поведение среднего оценок из первого пункта при росте размера $n$ выборки - постройте график зависимости среднего оценки по всем экспериментам от размера выборки. Для вычисления зависимости нужно один раз сгенерировать выборки из достаточно большого размера и посчитать оценки по префиксам, используя функции из `numpy`.

*Подсказка: кажется, на семинаре для этих целей использовалась функция `cumsum`...*

```python
max_sample_size = 500
n_runs = 100

np.random.seed(42)
X = norm(loc=0, scale=1).rvs(size=(n_runs, max_sample_size))
```

```python
cumsum_X = np.cumsum(X, axis=1)
cumsum_X2 = np.cumsum(X**2, axis=1)

n_values = np.arange(2, max_sample_size + 2)

var = (cumsum_X2 - (cumsum_X**2) / n_values) / n_values
var_corrected = var * (n_values / (n_values - 1))

mean_biased = np.mean(var, axis=0)
mean_unbiased = np.mean(var_corrected, axis=0)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(n_values, mean_biased, label=r'Смещенная оценка $S^2$', color='blue')
plt.plot(n_values, mean_unbiased, label=r'Несмещенная оценка $\frac{n}{n-1}S^2$', color='green')
plt.axhline(1, color='red', linestyle='--', label='Истинное значение $\sigma^2=1$')

plt.xlabel('Размер выборки $n$', fontsize=12)
plt.ylabel('Среднее оценок', fontsize=12)
plt.title('Зависимость среднего оценок дисперсии от размера выборки', fontsize=14)
plt.legend()
plt.grid(True)
plt.ylim(0.8, 1.2)
plt.show()
```

```python
theta_mnk = np.zeros((exp, d))
theta_ridge_1 = np.zeros((exp, d))
theta_ridge_2 = np.zeros((exp, d))

for i in range(exp):
    epsilon = np.random.normal(0, sigma, n)
    Y = X @ thetas + epsilon

    theta_mnk[i] = np.linalg.inv(X.T @ X) @ X.T @ Y

    theta_ridge_1[i] = np.linalg.inv(X.T @ X + lambdas[0] * np.eye(d)) @ X.T @ Y
    theta_ridge_2[i] = np.linalg.inv(X.T @ X + lambdas[1] * np.eye(d)) @ X.T @ Y
```

```python
for n in n_values:
    theta_mnk = []
    theta_ridge = []

    for _ in range(exp):
        X = np.random.randn(n, d)
        Q, _ = np.linalg.qr(X)
        X = Q[:, :d]

        epsilon = np.random.normal(0, sigma, n)
        Y = X @ thetas + epsilon

        theta_hat_mnk = np.linalg.inv(X.T @ X) @ X.T @ Y
        theta_mnk.append(theta_hat_mnk)

        theta_hat_ridge = np.linalg.inv(X.T @ X + lambdas * np.eye(d)) @ X.T @ Y
        theta_ridge.append(theta_hat_ridge)

    theta_mnk_means.append(np.mean(theta_mnk, axis=0))
    theta_ridge_means.append(np.mean(theta_ridge, axis=0))
```

```python
plt.figure(figsize=(12, 6))

for i in range(d):
    plt.subplot(1, 2, i + 1)

    plt.plot(n_values, [theta[i] for theta in theta_mnk_means],
             label="МНК", color="blue", linestyle="--")

    plt.plot(n_values, [theta[i] for theta in theta_ridge_means],
             label=f"Ridge (λ={lambdas})", color="red")

    plt.axhline(thetas[i], color="green", linestyle=":", label="Истинное значение")

    plt.xlabel("Размер выборки (n)", fontsize=12)
    plt.ylabel(f"Оценка θ_{i+1}", fontsize=12)
    plt.title(f"Зависимость оценки θ_{i+1} от размера выборки", fontsize=14)
    plt.grid(alpha=0.5)
    plt.legend()

plt.tight_layout()
plt.show()
```
